```python
#Se requiere calcular los productos con mayor venta (Agregar otra columna de transacción total)
cafe_df['transaction_total'] = cafe_df['transaction_qty'] * cafe_df['unit_price']
cafe_df.head(5)
```

```python
#Orden de mayor a menor
vta_producto = cafe_df.groupby('product_type')['transaction_total'].sum()
vta_producto_sorted = vta_producto.sort_values(ascending=False)
vta_producto_sorted
```

```python
# Desglosar la columna "transaction_date" en "Year", "Month" y "Day"
cafe_df['year'] = cafe_df['transaction_date'].dt.year
cafe_df['month'] = cafe_df['transaction_date'].dt.month_name()
cafe_df['day'] = cafe_df['transaction_date'].dt.day_name()
```

```python
#Función para graficar datos
def count_plot(x_axis = None, y_axis = None, hue = None, rotation = 0, top = None):
  if x_axis:
    order = cafe_df[x_axis].value_counts().iloc[:top].index
  else:
    order = cafe_df[y_axis].value_counts().iloc[:top].index

  graph = sns.countplot (x = x_axis, y = y_axis, data = cafe_df, hue = hue, order = order, palette = "RdBu")
  for container in graph.containers:
    graph.bar_label(container)

  plt.xticks(rotation = rotation)
  plt.show()
```

```python
#Mostramos la grafica de los meses con mayor número de ventas
count_plot(x_axis = "month")
```

```python
#Mostrar con la función los días con mayor número de ventas
count_plot(x_axis = "day")
```

```python
#Mostrar el horario en el que hay mayor demanda
cafe_df["hour"] = cafe_df["transaction_time"].apply(lambda x : x.hour)
sns.kdeplot(cafe_df["hour"])
plt.show()
```
